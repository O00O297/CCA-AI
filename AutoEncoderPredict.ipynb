```
from keras.layers import Dense, Flatten, Reshape, Input, InputLayer
from keras.models import Sequential, Model

def build_autoencoder(img_shape, code_size):
    # The encoder
    encoder = Sequential()
    encoder.add(InputLayer(img_shape))
    encoder.add(Flatten())
    encoder.add(Dense(code_size))

    # The decoder
    decoder = Sequential()
    decoder.add(InputLayer((code_size,)))
    decoder.add(Dense(np.prod(img_shape))) # np.prod(img_shape) is the same as 32*32*3, it's more generic than saying 3072
    decoder.add(Reshape(img_shape))

    return encoder, decoder

def getcode(img,encoder):
  code = encoder.predict(img[None])[0]
  return code

def load_Model():
  IMG_SHAPE = (128, 128)
  encoder, decoder = build_autoencoder(IMG_SHAPE,128)

  inp = Input(IMG_SHAPE)
  code = encoder(inp)
  reconstruction = decoder(code)

  autoencoder = Model(inp,reconstruction)
  autoencoder.load_weights('/content/drive/Shared drives/CCA-AI Slide/Dataset/Model/modelAutoEncoder1.h5')
  return autoencoder, encoder

def k_meanPredict(x_data):
  slide_dict = np.empty(0)
  for data in x_data:
    class_label =  np.empty(0)
    for centroid in arr :
      dst = distance.euclidean(centroid, data)
      class_label = np.append(class_label,dst)
    slide_dict = np.append(slide_dict, class_label.argmin())
  slide_dict = slide_dict.astype(np.int32)
  return slide_dict

def k_meanPredict1(x_data):
  class_label =  np.empty(0)
  for centroid in arr :
    dst = distance.euclidean(centroid, x_data)
    class_label = np.append(class_label,dst)
  slide_dict = class_label.argmin()
  slide_dict = slide_dict.astype(np.int32)
  return slide_dict

def histogramTiles(h_data, centroid):
  his = [0] * centroid
  for i in range(centroid):
    his[i] = h_data.count(i)
  return his
```
